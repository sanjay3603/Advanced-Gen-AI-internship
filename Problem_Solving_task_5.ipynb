```python
print("---Problem 1: Social Media – Post Engagement Analyzer---")

def PostLikes(likes):
  total=sum(likes)
  if total>=1000:
    status= "Viral Post"
  else:
    status= "Normal Engagement"
  print(f"Total likes:{total}")
  print(f"Post status:{status}")

likes=[100,200,500,700]
PostLikes(likes)
```

```python
print("---Problem 3. Agriculture – Rainfall Adequacy Checker---")


def RainfallChecker(rainfall,required_level):
  average=sum(rainfall)/len(rainfall)
  if average >=required_level:
    status="Adequate Rainfall"
  else:
    status="Inadequate Rainfall"
  print(f"Average Rainfall:{average}")
  print(f"Rainfall Status:{status}")
#user input
rainfall_input=input("Enter rainfall:")
rainfall = list(map(int, rainfall_input.split()))
required_level=int(input("Enter required level:"))
RainfallChecker(rainfall,required_level)
```

```python
print("---Problem 4.Social Media – Duplicate Account Detection---")


def dupliacteAccounts(username):
  unique_name=set(username)
  if len(unique_name)!=len(username):
    print("Duplicate Account Found:Yes")
  else:
    print("Duplicate Account Found:No")
#user input
user_input=input("Enter username: ")
username=user_input.split()
dupliacteAccounts(username)
```

```python
print("---Problem 6: Agriculture Premium Crop Filter---")


def premium_crop_filter(prices):
    premium_crops = []

    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)


# Example
premium_crop_filter([1500, 2500, 1800, 3200])
```

```python
print("---Problem 7.System Monitoring – Application Health Checker---")


def HealthChecker(errors):
  if errors ==0:
    status="Healthy"
  elif errors<=5:
    status="Minor Issues"
  else:
    status="Critical Issues"
  print(f"Error Count:{errors}")
  print(f"System Status:{status}")
#user input
errors=7
HealthChecker(errors)
```

```python
print("---Problem 8.Banking – Daily Transaction Limit Checker---")


def DailyTransaction(amount):
  Daily_limit=50000
  if amount<=Daily_limit:
    status="Approved"
  else:
    status="Rejected"
  print(f"Transaction Amount:{amount}")
  print(f"Transaction status:{status}")
#user input
amount=60000
DailyTransaction(amount)
```

```python
print("--- Problem 9.E-Learning – Student Attendance Eligibility System---")


def AttendanceCalculation(attendance):
  percentage=sum(attendance)/len(attendance)*100
  if percentage>=75:
    status="Eligible"
  else:
    status="Not Eligible"
  print(f"Attendance percentage:{percentage}")
  print(f"Exam Eligibilty:{status}")
#user input
attendance_input=input("Enter attendance(1 for Present, 0 for Absent):")
attendance=list(map(int,attendance_input.split()))
AttendanceCalculation(attendance)
```

```python
print("---Problem 10.Smart Electricity Bill Analyzer---")


def electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill:", bill)
    print("Usage Classification:", usage)


# Example
electricity_bill(250)
```
