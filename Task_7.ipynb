```python
print("---Problem 1: Smart Light Controller---\n")


class SmartLight:
    def __init__(self, light_name, status):
        self.light_name = light_name
        self.status = status

    def status_display(self):
        print(f"{self.light_name} is {self.status}")

smart_light = SmartLight("Bedroom Light", "ON")
smart_light.status_display()
```

```python
print("---Problem 2: Employee ID Card System---\n")



class EmployeeDetails:
    def __init__(self, employee_name, employee_ID, department):
        self.employee_name = employee_name
        self.employee_ID = employee_ID
        self.department = department

    def display_ID(self):
        print("Employee ID Card")
        print(f"Name: {self.employee_name}")
        print(f"ID: {self.employee_ID}")
        print(f"Department: {self.department}")

e = EmployeeDetails("Rahul", "EMP102", "IT")
e.display_ID()
```

```python
print("---Problem 3: Mobile Contact Record---\n")


class MobileContact:
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number
        print("Contact Saved")

    def display_contact(self):
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_number}")

anita = MobileContact("Anita", "9876543210")
anita.display_contact()
```

```python
print("---Problem 4: Product Price Tag Generator---\n")


class ProductPriceTag:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")

headphones = ProductPriceTag("Headphones", 2499)
headphones.print_tag()
```

```python
print("---Problem 5: Movie Rating Display System---\n")



class Movie:

    def __init__(self, movie_name, rating):
        # Store movie information
        self.movie_name = movie_name
        self.rating = rating

    def display_movie(self):

        print("\nMovie:", self.movie_name)
        print(f"Rating: {self.rating} / 5")


# Taking input
movie = input("Enter movie name: ")
rating = float(input("Enter movie rating: "))

movie1 = Movie(movie, rating)
movie1.display_movie()
```

```python
print("---Problem 6: Delivery Address Manager---\n")



class Delivery:

    def __init__(self, customer_name, address):
        # Store delivery details
        self.customer_name = customer_name
        self.address = address

    def display_delivery(self):

        print("\nDelivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Taking input
customer = input("Enter customer name: ")
address = input("Enter delivery address: ")

delivery1 = Delivery(customer, address)
delivery1.display_delivery()
```
