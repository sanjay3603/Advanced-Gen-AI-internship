```python
print("---Problem Statement 1:Smart Parking Lot Management System---\n")

def smart_parking_system(capacity, vehicle_logs):
    parked = 0
    peak = 0

    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Exceeded Capacity ")
    elif parked == capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")

smart_parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
print("---Problem Statement 2:Online Food Delivery Time Estimator---\n")

def estimate_delivery_time(distance, traffic, weather):
    time = distance * 5

    if traffic.lower() == "high":
        time += 15
    elif traffic.lower() == "medium":
        time += 10
    else:
        time += 5

    if weather.lower() == "rainy":
        time += 10
    elif weather.lower() == "storm":
        time += 20

    print("Estimated Delivery Time:", time, "minutes")

estimate_delivery_time(8, "High", "Rainy")
```

```python
print("---Problem Statement 3:Movie Theatre Seat Occupancy---\n")

def seat_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", round(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

seat_occupancy(200, [1]*150)
```

```python
print("---Problem Statement 4: Cloud Server Load Classification System---\n")



def load_classification(cpu_usage):
    # Calculate average
    total = 0
    for i in cpu_usage:
        total += i
    average = (total / len(cpu_usage)) # (the readings are already in %)

    # Check for average
    if average > 80:
        status = "Critical"
    elif average >= 50:
        status = "Warning"
    else:
        status = "Normal"

    return average, status

# main
CPUReadings = [45, 60, 70, 85, 90]
average_laod, server_status = load_classification(CPUReadings)
print(f"Average CPU Load: {average_laod:.0f}%")
print(f"Server Status: {server_status}")
```

```python
print("---Problem Statement 5: Smart Classroom Resource Usage Monitor---\n")



def resource_monitor(usage):

    # Create empty list
    overused = []

    # Check for overusage of resource
    for resource in usage:
        if usage[resource] > 8:
            overused.append(resource)

    # Condition for alert status
    if overused:
        alert = "Yes"
    else:
        overused.append("None")  # if no overuse, just gives None
        alert = "No"

    return ", ".join(overused), alert

# main
ResourceUsage = {
    "Projector" : 6,
    "AC" : 9,
    "Lights" : 4
}
overused_resources, energy_alert = resource_monitor(ResourceUsage)
print(f"Overused Rsources: {overused_resources}")
print(f"Energy Alert: {energy_alert}")
```

```python
print("---Problem Statement 6: Online Event Registration Capacity Controller---\n")



def event_manager(capacity, registrations):
    confirmed = 0
    waitlisted = 0

    # count confirmed and waitlisted users
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    # Condition to update status
    if confirmed < capacity :
        status = "Open"
    else:
        status = "Closed"

    return confirmed, waitlisted, status

# main
EventCapacity = 100
Registrations = 105
confirmed_registrations, waitlisted_users, registration_status = event_manager(EventCapacity, Registrations)
print(f"Confirmed Registrations: {confirmed_registrations}")
print(f"Waitlisted Users: {waitlisted_users}")
print(f"Registration Status: {registration_status}")
```
