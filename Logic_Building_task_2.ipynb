```python
print("---Problem 1: Unique Words in a Sentence---")


sentence = "python is easy and python is powerful"
a = sentence.split(' ')
b=set(a)
print("Unique words count:",len(b))
print("Unique words:",b)
```

```python
print("---Problem 2: Highest Salary from Employee Data---")


employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_name = None
highest_salary = 0
for name,salary in employees.items():
  if salary > highest_salary:
    highest_salary = salary
    highest_name = name
print("Highest Salary:",highest_name,"-",highest_salary)
```

```python
print("---Problem 3: Find Maximum and Minimum Values---")


numbers=[45,22,89,10,66]
min=numbers[0]
max=numbers[0]
for i in numbers:
    if i<min:
        min=i
    if i>max:
        max=i
print(f"List:{numbers}\nmax:{max}\nmin:{min}")
```

```python
print("---Problem 4: Count Products Above a Price Threshold---")


prices=[450,1200,899,1500,300]
margin=1000
count=0
for i in prices:
  if i > margin:
    count +=1
print(f"Products above 1000:{count}")
```

```python
print("---Problem 5: Calculate Attendance Percentage---")


attendence=["P","P","A","P","P"]
count=0
for i in attendence:
  if i == "P":
    count +=1
attendence_percentage=count/len(attendence)*100
print(f"Attendence Percentage:{attendence_percentage}")
```
